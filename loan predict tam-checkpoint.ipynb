```python
import numpy as np
import scipy as ap
import pandas as pd
```

```python
df = pd.read_csv('loan.csv')
```

```python
df = df.fillna(0)
df.head()
```

### assign the future into numpy np array

```python
x = df.values[0:,1:12]
```

```python
print(x[0])
```

### change string as numerical value

```python
j=0
for i in x:
    if i[0]=='Male':
        x[j,[0]]=1
    else:
        x[j,[0]]=0
        
    if i[1]=='Yes':
        x[j,[1]]=1
    else:
        x[j,[1]]=0
        
        
    if i[2]=='3+':
        x[j,[2]]=5
   
    if i[3]=='Graduate':
        x[j,[3]]=1
    else:
        x[j,[3]]= 0 
        
    if i[4]=='Yes':
        x[j,[4]]=1
    else:
        x[j,[4]]=0
        
        
    if i[10]=='Urban':
        x[j,[10]]=1
    elif i[10]=='Semiurban':
        x[j,[10]]=2
    else:
        x[j,[10]]=3   
        
        
        
    j=j+1
    
    
    
    
```

### define label in np array

```python
y = df.values[0:,12:]
```

### convert the string value to numerical value

```python
j=0
for i in y:
    if i[0]=='Y':
        y[j,[0]]=1
    else:
        y[j,[0]]=0
    j=j+1
```

### using sklearn linear regersion  for predict eligible
